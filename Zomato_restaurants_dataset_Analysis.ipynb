```python
import pandas as pd
```

```python
data = pd.read_csv('zomato.csv')
```

```python
data
```

```python
data.info()
```

```python
# 1. What is the average rating of restaurants for each type (e.g., Quick Bites, Cafe)?
```

```python
# Average rating by restaurant type
average_rating = data.groupby('restaurant type')['rate (out of 5)'].mean()
print(average_rating)
```

```python
# 2. Which area has the highest number of restaurants listed?
```

```python
# Count of restaurants by area
restaurant_count = data['area'].value_counts()
print(restaurant_count)
```

```python
# 3. What is the average cost for two people in restaurants that offer online ordering vs. those that don't?
```

```python
# Average cost comparison by online order availability
average_cost_online = data.groupby('online_order')['avg cost (two people)'].mean()
print(average_cost_online)
```

```python
# 4. Which restaurant(s) have the maximum number of ratings?
```

```python
# Restaurant(s) with the highest number of ratings
max_ratings = data['num of ratings'].max()
top_rated_restaurants = data[data['num of ratings'] == max_ratings]
print(top_rated_restaurants[['restaurant name', 'num of ratings']])
```

```python
# 5. What are the top 5 cuisines served most frequently in the dataset?
```

```python
# Top 5 most common cuisines
from collections import Counter

# Split cuisines and count frequency
cuisines_list = data['cuisines type'].str.split(',').sum()
top_cuisines = Counter(cuisines_list).most_common(5)
print(top_cuisines)
```
